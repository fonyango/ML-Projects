# Feature Engineering

```python
df['Avg. Area Number of Bedrooms'].value_counts()
```

```python
df=pd.get_dummies(df, columns=['Avg. Area Number of Bedrooms'])
```

```python
df.shape
```

# Creating and evaluating the model

```python
#Splittig the data set

y = df['Price']

x = df.drop('Price', axis=1)
x = df.drop('Address', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
#Training the model

from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
print('Coefficients: \n', lm.coef_)
```

```python
#Predicting test data

predictions = lm.predict(x_test)
```

```python
#creating a scatter plot of the real values vs predicted values

plt.scatter(y_test, predictions)
plt.xlabel('Y test')
plt.ylabel('Prediction Y')
```

```python
#Evaluating the model

from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test,predictions))
print('MSE:', metrics.mean_squared_error(y_test,predictions))
print('RMSE:', np.sqrt(metrics.mean_absolute_error(y_test,predictions)))
```

```python
#Plotting a histogram for the residuals

sns.distplot((y_test-predictions), bins=50)
```

```python
#Interpreting coefficients

coefficients = pd.DataFrame(lm.coef_, x.columns)
coefficients.columns = ['Coefficients']
coefficients
```
